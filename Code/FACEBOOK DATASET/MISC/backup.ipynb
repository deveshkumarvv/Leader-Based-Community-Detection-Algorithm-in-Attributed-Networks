```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import community     # Import the community module
import random  # Import the random module
```

```python
file_path = 'facebook_combined.txt'
```

```python
# Read an edge list from a file specified by 'file_path'
# Create a graph (G) where nodes are of integer type
# This assumes the edge list file contains pairs of integers representing edges

G = nx.read_edgelist(file_path, nodetype=int)
```

```python
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

```python
# Create a spring layout for the graph
pos = nx.spring_layout(G)
```

```python
# Draw the graph
nx.draw(G, pos, with_labels=False, node_size=9, node_color="green", edge_color="black")
```

```python
nx.draw_networkx_edges(G, pos, edge_color='black')# showing the edges 
```

```python
for node in G.nodes():
    G.nodes[node]['score1'] = random.randint(1, 1000)
    G.nodes[node]['score2'] = random.randint(1, 1000)
    G.nodes[node]['score3'] = random.randint(1, 1000)
```

```python
# Print the attributes of the first few nodes as a check
for node, data in G.nodes(data=True):
    print(f"Node {node}: {data}")
```

```python
# Find the top 10 leaders based on 'score1'
top_leaders_score1 = sorted(G.nodes(data=True), key=lambda x: x[1]['score1'], reverse=True)[:10]
```

```python
# Extract node IDs from the result
top_leaders_ids_score1 = [node_id for node_id, _ in top_leaders_score1]
```

```python
# Print or use the top 10 leaders based on 'score1'
print("Top 10 leaders based on score1:", top_leaders_ids_score1)
```

```python
# Display the top 10 leaders with their scores
print("Top 10 leaders based on score1:")
for rank, (node_id, data) in enumerate(top_leaders_score1, start=1):
    score1 = data['score1']
    print(f"{rank}. Node {node_id}: Score1 = {score1}")
```

```python
# Define a function to calculate the leadership score based on the three attributes
def calculate_leadership_score(node_data):
    return (node_data['score1'] + node_data['score2'] + node_data['score3']) / 3
```

```python
# Add a 'leadership_score' attribute to each node
for node, data in G.nodes(data=True):
    G.nodes[node]['leadership_score'] = calculate_leadership_score(data)
```

```python
# Print the attributes of the first few nodes as a check
for node, data in G.nodes(data=True):
    print(f"Node {node}: {data}")
```

```python
# Find the top 13 leaders based on the 'leadership_score'
top_leaders = sorted(G.nodes(data=True), key=lambda x: x[1]['leadership_score'], reverse=True)[:13]
```

```python
# Print the top leaders
print("Top 13 Leaders:")
for node, data in top_leaders:
    print(f"Node {node}: Leadership Score {data['leadership_score']}")
```

```python
# Calculate eigenvector centrality scores
eigenvector_centrality_scores = nx.eigenvector_centrality(G)
```

```python
# Add 'eigenvector_centrality' attribute to each node
for node, score in eigenvector_centrality_scores.items():
    G.nodes[node]['eigenvector_centrality'] = score
```

```python
# Find the top 13 leaders based on a combination of 'leadership_score' and 'eigenvector_centrality'
combined_scores = [
    (node, data['leadership_score'] + data['eigenvector_centrality'])
    for node, data in G.nodes(data=True)
]
```

```python
# Select the top 13 nodes based on the combined scores
top_leaders = sorted(combined_scores, key=lambda x: x[1], reverse=True)[:13]
```

```python
# Print the top leaders
print("Top 13 Leaders:")
for node, combined_score in top_leaders:
    print(f"Node {node}: Combined Score {combined_score} (Leadership Score {G.nodes[node]['leadership_score']}, Eigenvector Centrality {G.nodes[node]['eigenvector_centrality']})")
```

```python
# Perform community detection ensuring each leader is in a separate community
communities = list(nx.community.greedy_modularity_communities(G))
```

```python
# Create a dictionary to store community assignments for each leader
leader_communities = {leader: i for i, leader in enumerate(top_leaders)}
```

```python
# Create a dictionary to store the final community assignments
community_assignments = {node: -1 for node in G.nodes()}
```

```python
# Assign nodes to communities
for i, community_set in enumerate(communities):
    for node in community_set:
        if node in top_leaders:
            community_assignments[node] = leader_communities[node]
        else:
            community_assignments[node] = i + len(top_leaders)
```

```python
# Create a spring layout for the graph
pos = nx.spring_layout(G)
```

```python
# Draw the graph with node colors based on the community assignments
node_colors = [community_assignments[node] for node in G.nodes()]
nx.draw(G, pos, with_labels=False, node_size=10, node_color=node_colors, cmap=plt.cm.get_cmap('viridis'))
```

```python
# Add community information as node attributes
for node, community in community_assignments.items():
    G.nodes[node]['community'] = community

# Export the graph to a GEXF file
nx.write_gexf(G, "graph_with_communities.gexf")
```

```python
num_communities = len(communities)
print(f"The number of communities is: {num_communities}")
```
