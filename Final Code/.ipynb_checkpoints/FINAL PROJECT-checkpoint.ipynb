```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import community     # Import the community module
import random  # Import the random module
```

```python
file_path = 'facebook_combined.txt'
```

```python
# Read an edge list from a file specified by 'file_path'
# Create a graph (G) where nodes are of integer type
# This assumes the edge list file contains pairs of integers representing edges

G = nx.read_edgelist(file_path, nodetype=int)
```

```python
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

```python
nodes_to_keep = list(G.nodes)[:2000] #keeping the 2000 nodes
```

```python
# Remove nodes that are not in the first 2000
nodes_to_remove = set(G.nodes) - set(nodes_to_keep)
G.remove_nodes_from(nodes_to_remove)
```

```python
pos = nx.spring_layout(G)  # You can choose a different layout algorithm
nx.draw(G, pos, with_labels=False, node_size=10, node_color='green', alpha=0.5)
```

```python
nx.draw_networkx_edges(G, pos, edge_color='gray')# showing the edges 
```

```python
nx.draw_networkx_labels(G, pos, font_color='black', font_size=10)
# Draw labels on the networkx graph 'G' at specified positions 'pos'
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G)
```

```python
leader = max(eigenvector_centrality, key=eigenvector_centrality.get)
```

```python
print("Leader node:", leader)
```

```python
top_nodes = sorted(eigenvector_centrality, key=eigenvector_centrality.get, reverse=True)[:10]
```

```python
print("Top 10 nodes based on eigenvector centrality:", top_nodes)
```

```python
G_top = G.subgraph(top_nodes)
```

```python
pos = nx.spring_layout(G_top)  
```

```python
nx.draw(G_top, pos, with_labels=True, font_weight='bold')
```

```python
 # Use Louvain method for community detection
partition = community.best_partition(G)
```

```python
# Extract unique community IDs
unique_communities = set(partition.values())
```

```python
# Define 10 distinct colors for the 10 communities
community_colors = [
    'red', 'blue', 'green', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'brown', 'gray'
]
```

```python
# Map community IDs to colors
community_color_map = {community_id: community_colors[i % len(community_colors)] for i, community_id in enumerate(unique_communities)}
```

```python
# Create a list of colors for each node based on its community
node_colors = [community_color_map[partition[node]] for node in G.nodes]
```

```python
# Plot the graph with nodes colored by community, alternative labels, smaller font size, and smaller node size
pos = nx.spring_layout(G)
```

```python
# Create a mapping of original node labels to smaller labels
label_mapping = {node: str(i) for i, node in enumerate(G.nodes)}
```

```python
# Specify the font size and node size
font_size = 6
node_size = 80
```

```python
# Draw the graph with alternative labels, smaller font size, smaller node size, and distinct community colors
nx.draw(G, pos, node_color=node_colors, with_labels=True, labels=label_mapping, font_size=font_size, node_size=node_size)
```

```python
# Add random attributes for 100 nodes
for node_id in range(1, 101):
    communication_skills = random.randint(1, 10)
    initiative = random.randint(1, 10)
    decision_making = random.randint(1, 10)

    G.add_node(
        node_id,
        communication_skills=communication_skills,
        initiative=initiative,
        decision_making=decision_making
    )
```

```python
print(G.nodes(data=True))
```

```python
# Calculate a hypothetical leadership score based on attributes
for node_id in G.nodes:
    communication_skills = G.nodes[node_id].get('communication_skills', 0)
    initiative = G.nodes[node_id].get('initiative', 0)
    decision_making = G.nodes[node_id].get('decision_making', 0)

    leadership_score = communication_skills + initiative + decision_making
    G.nodes[node_id]['leadership_score'] = leadership_score
```

```python
print(G.nodes(data=True))# To view leadership scores of each nodes
```

```python
# Find the leader based on the best communication skills
leader = max(G.nodes, key=lambda node: G.nodes[node].get('communication_skills', 0))

print(f"The leader is Node {leader} with the best communication skills: {G.nodes[leader]['communication_skills']}.")
```

```python
# Find the top 15 leaders based on communication skills
top_leaders = sorted(G.nodes, key=lambda node: G.nodes[node].get('communication_skills', 0), reverse=True)[:15]

print("Top 15 leaders based on communication skills:")
for leader in top_leaders:
    print(f"Node {leader} - Communication Skills: {G.nodes[leader]['communication_skills']}")
```

# Visualize the graph with the top leaders highlighted


```python
pos = nx.spring_layout(G)  # We can choose a different layout algorithm
nx.draw(G, pos, with_labels=False, node_size=50, alpha=0.7)
```

```python
# Highlight the top leaders with a different color
node_colors = ['red' if node in top_leaders else 'skyblue' for node in G.nodes]
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=50)
```

```python
# Display labels for the top leaders
labels = {node: f"{node}\nComm Skills: {G.nodes[node]['communication_skills']}" for node in top_leaders}
nx.draw_networkx_labels(G, pos, labels=labels, font_color='black', font_size=8)
```

```python
# Find the top 10 leaders based on the leadership score
top_leaders = sorted(G.nodes, key=lambda node: G.nodes[node].get('leadership_score', 0), reverse=True)[:10]
```

```python
print("Top 10 leaders based on the attributes:")
for leader in top_leaders:
    print(f"Node {leader} - Leadership Score: {G.nodes[leader]['leadership_score']}")
```

```python
# Calculate eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G)
```

```python
# Combine leadership score and eigenvector centrality for prioritization
priority_scores = {node: (G.nodes[node]['leadership_score'], eigenvector_centrality[node]) for node in G.nodes}
```

```python
# Find the top 10 leaders based on priority scores
top_leaders = sorted(G.nodes, key=lambda node: priority_scores[node], reverse=True)[:10]
```

```python
print("Top 10 leaders based on priority scores:")
for leader in top_leaders:
    print(f"Node {leader} - Leadership Score: {G.nodes[leader]['leadership_score']}, Eigenvector Centrality: {eigenvector_centrality[leader]}")
```

```python
# Create a subgraph containing only the top leaders
leader_subgraph = G.subgraph(top_leaders)
```

```python
# Visualize the subgraph of top leaders
pos = nx.spring_layout(G)  # You can choose a different layout algorithm
nx.draw(leader_subgraph, pos, with_labels=True, node_size=200, node_color='red', font_color='black', font_size=8)
```

```python
# Visualize the entire graph
pos = nx.spring_layout(G)  # You can choose a different layout algorithm
nx.draw(G, pos, with_labels=False, node_size=50, node_color='skyblue', alpha=0.7)
```

```python
# Use Louvain method for community detection
partition = community.best_partition(G)
```

```python
# Extract the top 10 communities
top_communities = sorted(set(partition.values()), key=lambda comm: list(partition.values()).count(comm), reverse=True)[:10]
```

```python
# Visualize the entire graph with edges and 10 colors for 10 communities
pos = nx.spring_layout(G)  # You can choose a different layout algorithm
```

```python
# Plot nodes for each community with a different color
for i, community_id in enumerate(top_communities):
    community_nodes = [node for node, comm_id in partition.items() if comm_id == community_id]
    nx.draw_networkx_nodes(G, pos, nodelist=community_nodes, node_color=plt.cm.tab10(i), node_size=20)
```

```python
# Add node labels
labels = {node: str(node) for node in G.nodes}
nx.draw_networkx_labels(G, pos, labels, font_size=8)
```
